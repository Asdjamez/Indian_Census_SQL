Population of India

```python
%%sql
select sum(population) as total_population from Census2;
```

Average Growth Percentage of India

```python
%%sql
select round((avg(growth)*100),2) as avg_growth from Census1;
```

Average Growth Percentage of Indian States

```python
%%sql 
select state, round((avg(growth)*100),2) as avg_growth from Census1 group by State;
```

Average Sex Ratio

```python
%%sql
select state, round((avg(Sex_Ratio)),0) as avg_sex_ratio from Census1 group by State order by 2 desc;
```

Average Literacy Rate more than 90

```python
%%sql
select state, round((avg(Literacy)),0) as avg_literacy from Census1 
group by State having round((avg(Literacy)),0) >90 order by 2 desc
```

Top 3 states having highest growth ratio

```python
%%sql 
select state, round((avg(Growth)*100),0) as avg_growth from Census1 
group by State order by 2 desc limit 3;
```

Bottom 3 states having lowest sex ratio

```python
%%sql
select state, round((avg(Sex_Ratio)),0) as avg_sex_ratio from Census1 
group by State order by 2 Limit 3;
```

Top and Bottom 3 states in literacy

```python
%%sql
with topstates as(
select state, round((avg(Literacy)),0) as avg_literacy from Census1 
group by State order by avg_literacy desc)

select * from topstates limit 3;
```

```python
%%sql 
with bottomstates as
(select state, round((avg(Literacy)),0) as avg_literacy from Census1 
group by State order by avg_literacy asc)

select * from bottomstates limit 3;
```

States starting with letter a

```python
%%sql
select distinct state from Census1 where LOWER(state) like 'a%'
```
